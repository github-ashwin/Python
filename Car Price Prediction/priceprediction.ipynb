```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('./car_price.csv')
df
```

```python
df.info()
```

```python
columns = ['car_prices_in_rupee','kms_driven','fuel_type','transmission','ownership','manufacture','engine','Seats']
df = df[columns]
df
```

```python
df['kms_driven'] = df['kms_driven'].str.split().str[0].str.replace(',','').astype(int)
df['kms_driven']
```

```python
df['fuel_type'].unique()
```

```python
label_encoder = LabelEncoder()
df['fuel_type']= label_encoder.fit_transform(df['fuel_type'])
df['fuel_type']
```

```python
df['transmission'].unique()
```

```python
df['transmission']= label_encoder.fit_transform(df['transmission'])
df['transmission']
```

```python
df['ownership'].unique()
```

```python
# ownership = df['ownership'].str.split().str[0].str[0].astype(int)
# ownership
df['ownership'] = df['ownership'].str.split().str[0].str[0].astype(int)
df['ownership']
```

```python
df.info()
```

```python
df['engine'] = df['engine'].str.split().str[0].astype(int)
df['engine']
```

```python
df
```

```python
df['Seats'].unique()
```

```python
df['Seats'] = df['Seats'].str.split().str[0].astype(int)
df['Seats']
```

```python
df
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].str.split().str[0].str.replace(',','').astype(float)
df['car_prices_in_rupee']
```

```python
df.info()
```

```python
x = df[['kms_driven','fuel_type','transmission','ownership','manufacture','engine','Seats']]
x
```

```python
y = df[['car_prices_in_rupee']]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
prediction = lr.predict(x_test)
prediction
```

```python
print('Mean Absolute Error',mean_absolute_error(y_test,prediction))
print('Mean Squared Error',mean_squared_error(y_test,prediction))
print('R2 Score',r2_score(y_test,prediction))
```
