```python
df.drop(['Route'],axis=1,inplace=True)
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop(['Date_of_Journey'],axis=1,inplace=True)
df
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
df['Arrival_hr'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df
```

```python
df.drop(['Arrival_Time'],axis=1,inplace=True)
df
```

```python
df['Arrival_hr'] = df['Arrival_hr'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
df.info()
```

```python
df['Dep_Hr'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Min'] = df['Dep_Time'].str.split(':').str[1]
df
```

```python
df.drop(['Dep_Time'],axis=1,inplace=True)
df
```

```python
df['Dep_Hr'] = df['Dep_Hr'].astype(int)
df['Dep_Min'] = df['Dep_Min'].astype(int)
df.info()
```

```python
df
```

```python
df['Duration_hr'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df
```

```python
df.drop(['Duration'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder  = LabelEncoder()
source_label  = labelencoder.fit_transform(df['Source'])
destination_label = labelencoder.fit_transform(df['Destination'])

print(source_label)
print()
print(destination_label)
```

```python
df['Source_label'] = source_label
df['Destination_label'] = destination_label

df
```
