```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html.parser')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
table_title = table.find_all('th')
```

```python
titles = [title.text.strip() for title in table_title]
titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=titles)
df
```

```python
col_data = table.find_all('tr')
col_data
```

```python
for row in col_data:
    row_data = row.find_all('td')
    item = [td.get_text(strip=True).replace("\n", "") for td in row_data]
    
    # print(item)
    if len(item) == len(titles):
        df.loc[len(df)] = item
df
```
